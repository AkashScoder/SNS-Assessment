```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained model and tokenizer
model_name = 'gpt2-large'  #Model
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Function to generate text based on a prompt
def generate_text(prompt, max_length=100, temperature=0.7, top_k=50, top_p=0.9, repetition_penalty=1.2):
    # Tokenize input prompt
    inputs = tokenizer(prompt, return_tensors='pt')

    # Generate text with improved control
    outputs = model.generate(
        input_ids=inputs['input_ids'],
        attention_mask=inputs['attention_mask'],
        max_length=max_length,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        temperature=temperature,
        top_k=top_k,  # Limit sampling pool to the top k tokens
        top_p=top_p,  # Cumulative probability for nucleus sampling
        repetition_penalty=repetition_penalty,  # Penalize repeated tokens
        do_sample=True  # Enable sampling
    )

    # Decode and return generated text
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return generated_text

# Example usage
if __name__ == "__main__":
    prompt = input('Enter the prompt: ')
    generated_text = generate_text(prompt)

    print("Generated Text:")
    for line in generated_text.split('.'):
        if line.strip():
            print(line.strip() + '.')
```
